# ££ Split dataset

```python
y=df.iloc[:,0:1]
x=df.iloc[:,1:]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x=pd.DataFrame(x)
x=x.set_axis(['koi_score', 'koi_fpflag_nt', 'koi_fpflag_ss',
       'koi_fpflag_co', 'koi_fpflag_ec', 'koi_period', 'koi_time0bk',
       'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq',
       'koi_insol', 'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff',
       'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

# ££ LOgistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred=lr.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
accuracy_score(ytest,ypred)
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(cm,cmap="Blues")
```

# KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
Knn=KNeighborsClassifier(n_neighbors=3)
Knn.fit(xtrain,ytrain)
ypred=Knn.predict(xtest)

print(accuracy_score(ytest,ypred),'/n',classification_report(ytest,ypred))
```

```python
print()
```

# Hypertuning KNN

```python
lost=[]
for i in range(1,11):
    Knn=KNeighborsClassifier(n_neighbors=i)
    Knn.fit(xtrain,ytrain)
    ypred=Knn.predict(xtest)

    lost.append(accuracy_score(ytest,ypred))
lost
```

# AFTER Hypertuning best k value

```python
from sklearn.neighbors import KNeighborsClassifier
Knn=KNeighborsClassifier(n_neighbors=22)
Knn.fit(xtrain,ytrain)
ypred=Knn.predict(xtest)

print(accuracy_score(ytest,ypred),'\n',"\n",confusion_matrix(ytest,ypred))
```
